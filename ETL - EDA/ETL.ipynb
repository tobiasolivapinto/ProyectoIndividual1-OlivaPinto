```python
#Importamos librerias a utilizar
import pandas as pd
```

```python
#importamos la base de datos y visualizamos
movies_db = pd.read_csv("Bases de Datos/movies_dataset.csv")
movies_db.head()
```

Consigna
- "Eliminar las columnas que no serán utilizadas, video,imdb_id,adult,original_title,vote_count,poster_path y homepage."

```python
#Importamos la libreria ast, que nos ayudara a desanidar las columnas que lo necesitan. 
import ast
```

```python
#Creamos la funcion para desanidar el primer caso
def fetch_name2(obj): 
    
    if isinstance(obj, str) and '{' in obj:
        # print(obj)
        dic = ast.literal_eval(obj)

        return dic['name']
```

```python
#Creamos la funcion para desanidar los otros casos
def fetch_name4(obj):
    if isinstance(obj, str) and '{' in obj:
        L=[]
        for i in ast.literal_eval(obj):
            L.append(i['name'])
        return L
    else:
        return []
```

```python
#Se lo aplicamos a las columnas correspondientes
movies_db['belongs_to_collection'] = movies_db['belongs_to_collection'].apply(fetch_name2)
```

```python
#Se lo aplicamos a las columnas correspondientes
movies_db['production_companies']  = movies_db['production_companies'].apply(fetch_name4)
```

```python
movies_db['production_countries']  = movies_db['production_countries'].apply(fetch_name4)
```

```python
movies_db['genres']  = movies_db['genres'].apply(fetch_name4)
```

```python
movies_db['spoken_languages']  = movies_db['spoken_languages'].apply(fetch_name4)
```
